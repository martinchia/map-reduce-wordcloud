```python
def co_occurrence(fpairs):
    dic = {}
    for line in fpairs:
        pairs, count = line.strip().split('\t')
        pairs = pairs.split(',')
        try:
            dic[pairs[0]][pairs[1]] = int(count)
        except KeyError:
            dic[pairs[0]] = {pairs[1]:int(count)}
    return dic
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def wordcloudPlt(frequency,title):
    wordcloud = WordCloud(max_words=1628,relative_scaling=1,normalize_plurals=False,
                          background_color='white',max_font_size=40,
                          ).generate_from_frequencies(frequency)
    plt.figure(1, figsize=(10,10))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
fNews = open("NewsCoOutput/part-00000",'r')
fTwitter = open("TwitterCoOutput/part-00000",'r')
dicNews = co_occurrence(fNews)
dicTwitter = co_occurrence(fTwitter)
fNews.close()
fTwitter.close()
```

```python
for k in dicNews:
    wordcloudPlt(dicNews[k],f"word co-occurrence of {k} for NYTimes")
```

```python
for k in dicTwitter:
    wordcloudPlt(dicTwitter[k],f"word co-occurrence of {k} for Twitter")
```
